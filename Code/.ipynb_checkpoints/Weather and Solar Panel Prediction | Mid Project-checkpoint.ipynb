### Data Used
***

```python
weather = pd.read_csv('weather_data.csv')
weather['cloudcover'] = weather['cloudcover'].fillna(0)
energy = pd.read_csv('energy_data.csv',sep=';')
```

## Steps

## Data Cleaning
***

#### Droping useless columns

```python
weather.drop(['solarradiation','solarenergy','uvindex'],axis=1,inplace=True)
```

#### Changing a colum error, podgoica to Podgorica

```python
weather['name'] = weather['name'].replace('podgoica', 'podgorica')
```

#### Separating datetime by day, month, year:

```python
weather['datetime'] = pd.to_datetime(weather['datetime'],errors='coerce')
weather['day'] = weather['datetime'].dt.day
weather['month'] = weather['datetime'].dt.month
weather['year'] = weather['datetime'].dt.year
```

#### Adding a column called day_duration

```python
weather['sunrise'] = pd.to_datetime(weather['sunrise'],errors='coerce')
weather['sunset'] = pd.to_datetime(weather['sunset'],errors='coerce')
weather["hours_of_sun"] = weather["sunset"] - weather["sunrise"]
weather['hours_of_sun'] = weather['hours_of_sun'].dt.total_seconds() / 3600
weather['sunrise'] = weather['sunrise'].dt.time
weather['sunset'] = weather['sunset'].dt.time
weather.drop(['datetime'],axis=1,inplace=True)
weather.drop(['sunrise'],axis=1,inplace=True)
weather.drop(['sunset'],axis=1,inplace=True)
```

#### Changing name column to city

```python
weather = weather.rename(columns={'name':'city'})
```

#### Reorganizing columns 

```python
weather = weather[['city','day','month','year','hours_of_sun', 'latitude', 'longitude', 'temp', 'humidity',
       'precip', 'windspeed', 'cloudcover']]
```

#### Deleting the year 2023 becouse is useless 

```python
weather = weather.drop(weather[weather.year == 2023].index)
```

#### Creating a table just with the averages per month

```python
clean_weather = weather.groupby(['city', 'month']).agg({'hours_of_sun': 'sum','latitude':'mean','longitude':'mean', 'temp': 'mean', 'humidity': 'mean', 'precip': 'mean', 'windspeed': 'mean', 'cloudcover': 'mean'}).reset_index()
clean_weather = clean_weather[['city', 'month', 'latitude', 'longitude', 'hours_of_sun', 'temp',
      'humidity', 'precip', 'windspeed', 'cloudcover']]
```

#### Merge weather data with energy data

```python
data = pd.merge(clean_weather, energy, on=['city', 'month'])
```

#### Creating a column with the energy produced per month

```python
data['monthly_energy_produced_kwh'] = (data['hours_of_sun']) * (data['solar_panel_power_kwh'])*(1 - (data['cloudcover']/100))*(1 - (data['precip'] / 100))/1000
```

#### Creating a column with the Montlhy_energy_cost

```python
data['montlhy_energy_cost_eur'] = data['avg_mth_en_consumption_house_kwh']*data['energy_price_eur_kwh']
```

#### Creating a column with the Montlhy_solar_panel_cost_saved

```python
data['montlhy_solar_panel_cost_saved'] = data['monthly_energy_produced_kwh']*data['energy_price_eur_kwh']
```

#### Creating a column with the money_saved

```python
data['money_saved'] = round((data['montlhy_solar_panel_cost_saved']/data['montlhy_energy_cost_eur'])*100,0)
data
```

### Droping useless columns

```python
data = data.drop(['windspeed','solar_panel_brand','solar_panel_efficiency_%'],axis=1)
```

## Testing my data
***

```python
label_encoder = LabelEncoder()
encoded_city = label_encoder.fit_transform(data['city'])
data['encoded_city'] = encoded_city
numerical_data=data.drop(['city'],axis=1)
```

```python
X = numerical_data.drop(['latitude','longitude'], axis=1)
Y = numerical_data[['latitude','longitude']]
X = sm.add_constant(X)
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.40,random_state=150)
model = sm.OLS(Y_train,X_train).fit()
predictions = model.predict(X_test)
r2_score(Y_test,predictions)
```

#### Deleting encoded_city

```python
data.drop(['encoded_city'],axis=1,inplace=True)
```

#### Creating a column with solar panel price

```python
data['solar_panel_price'] = 700
```

#### Creating a column with months_to_pay_the_solar_panel

```python
data['years_to_pay_the_solar_panel'] = (data['solar_panel_price']/data['montlhy_solar_panel_cost_saved'])/12
```

```python
data.to_csv('data.csv')
```

## Create a function that gives me to all cities in Europe
***

```python
def predict_weather_for_city(final_data):
    geolocator = Nominatim(user_agent="my_app")
    city = input("Enter the name of the city: ")
    location = geolocator.geocode(city)
    lat = location.latitude
    lon = location.longitude

    new_city = [(lat, lon)]
    distances = cdist(final_data[["latitude", "longitude"]], new_city, metric="euclidean")

    accuracies = []
    for k in range(1, 21):
        nearest_neighbor_indices = distances.argsort(axis=0)[:k, 0]
        actual_month = final_data.iloc[nearest_neighbor_indices]["month"].mode()[0]
        predicted_month = final_data.iloc[nearest_neighbor_indices]["month"].mean()
        accuracy = 100 - abs((predicted_month - actual_month) / actual_month) * 100
        accuracies.append(accuracy)
    print("Accuracies for different values of k:")
    for k, accuracy in enumerate(accuracies):
        print(f"k={k+1}: {accuracy:.2f}%")

    predicted_month = final_data.iloc[nearest_neighbor_indices]["month"].mean()
    predicted_hours_of_sun = final_data.iloc[nearest_neighbor_indices]["hours_of_sun"].mean()
    predicted_temperature = final_data.iloc[nearest_neighbor_indices]["temp"].mean()
    predicted_humidity = final_data.iloc[nearest_neighbor_indices]["humidity"].mean()
    predicted_precip = final_data.iloc[nearest_neighbor_indices]["precip"].mean()
    predicted_cloudcover = final_data.iloc[nearest_neighbor_indices]["cloudcover"].mean()
    predicted_avg_mth_en_consumption_house_kwh = final_data.iloc[nearest_neighbor_indices]["avg_mth_en_consumption_house_kwh"].mean()
    predicted_energy_price_eur_kwh = final_data.iloc[nearest_neighbor_indices]["energy_price_eur_kwh"].mean()
    predicted_solar_panel_power_kwh = final_data.iloc[nearest_neighbor_indices]["solar_panel_power_kwh"].mean()
    predicted_solar_panel_price = final_data.iloc[nearest_neighbor_indices]["solar_panel_price"].mean()

    print("\nPredicted weather conditions for the new city:")
    print("Hours of Sun:" , predicted_hours_of_sun)
    print("Temperature:", predicted_temperature)
    print("Humidity:", predicted_humidity)
    print("Precipitation:", predicted_precip)
    print("Cloud Cover:", predicted_cloudcover)
    print("avg_mth_en_consumption_house_kwh:", predicted_avg_mth_en_consumption_house_kwh)
    print("energy_price_eur_kwh:", predicted_energy_price_eur_kwh)
    print("solar_panel_power_kwh:", predicted_solar_panel_power_kwh)

    new_city_data = {
        "city": city,
        "month": predicted_month,
        "latitude": lat,
        "longitude": lon,
        "hours_of_sun":predicted_hours_of_sun,
        "temp": predicted_temperature,
        "humidity": predicted_humidity,
        "precip": predicted_precip,
        "cloudcover": predicted_cloudcover,
        "avg_mth_en_consumption_house_kwh":predicted_avg_mth_en_consumption_house_kwh,
        "energy_price_eur_kwh":predicted_energy_price_eur_kwh,
        "solar_panel_power_kwh":predicted_solar_panel_power_kwh,
        "monthly_energy_produced_kwh":(predicted_hours_of_sun * predicted_solar_panel_power_kwh *(1 - (predicted_cloudcover / 100)))/1000,
        "montlhy_energy_cost_eur": predicted_avg_mth_en_consumption_house_kwh * predicted_energy_price_eur_kwh,
        "montlhy_solar_panel_cost_saved":((predicted_hours_of_sun * predicted_solar_panel_power_kwh *(1 - (predicted_cloudcover / 100)))/1000)*(predicted_energy_price_eur_kwh),
        "money_saved": round(((((predicted_hours_of_sun * predicted_solar_panel_power_kwh *(1 - (predicted_cloudcover / 100)))/1000)*(predicted_energy_price_eur_kwh))/(predicted_avg_mth_en_consumption_house_kwh * predicted_energy_price_eur_kwh))*100,0),
        "solar_panel_price":predicted_solar_panel_price,
        "years_to_pay_the_solar_panel":(predicted_solar_panel_price/(((predicted_hours_of_sun * predicted_solar_panel_power_kwh *(1 - (predicted_cloudcover / 100)))/1000)*(predicted_energy_price_eur_kwh)))/12
    }
    final_data = final_data.append(new_city_data, ignore_index=True)

    return final_data
```

```python
final_data = predict_weather_for_city(final_data)
final_data
```
